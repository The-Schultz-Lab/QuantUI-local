## Introduction: What is Multiplicity?

### Electron Spin
Every electron has **spin** - a quantum property that can be:
- **↑** (spin-up, α, +½)
- **↓** (spin-down, β, -½)

### Pairing
- **Paired electrons:** ↑↓ (opposite spins)
- **Unpaired electrons:** ↑ or ↓ (no partner)

### Multiplicity Formula
$$\text{Multiplicity} = 2S + 1$$

Where $S$ is total spin:
$$S = \frac{n_\alpha - n_\beta}{2}$$

- $n_\alpha$ = number of spin-up electrons
- $n_\beta$ = number of spin-down electrons

### Common Cases
| System | Unpaired | $S$ | Multiplicity | Name |
|--------|----------|-----|--------------|------|
| H₂O | 0  | 0 | 1 | Singlet |
| O₂ (ground) | 2 | 1 | 3 | Triplet |
| Radical (·OH) | 1 | ½ | 2 | Doublet |
| CH₂ (ground) | 2 | 1 | 3 | Triplet |

---

## The Oxygen Paradox

### Experimental Fact
Molecular oxygen (O₂) is **magnetic** - it's attracted to magnets!

### Explanation
- Ground state O₂ has **2 unpaired electrons**
- Triplet state: ↑ ↑ (parallel spins)
- **Not** singlet: ↑↓ (paired)

### Why Does This Matter?
- Wrong multiplicity → Wrong energy (often by a lot!)
- Wrong reactivity predictions
- Wrong molecular properties

## Step 3: Understanding Singlet vs. Triplet

### Singlet O₂ (Multiplicity = 1)
- **All electrons paired:** ↑↓ ↑↓ ↑↓ ... ↑↓
- **Diamagnetic** (not magnetic)
- **Higher energy** (excited state)
- $S = 0$, so $2S+1 = 1$

### Triplet O₂ (Multiplicity = 3)
- **2 unpaired electrons:** ...↑↓ ↑↓ ↑↓ ↑↓ ↑ ↑
- **Paramagnetic** (magnetic!)
- **Lower energy** (ground state)
- $S = 1$, so $2S+1 = 3$

### Run the cell below ⬇️

```python
# Define both multiplicities
charge = 0  # Neutral molecule

# Singlet configuration
mult_singlet = 1
print("1️⃣  SINGLET O₂ (Multiplicity = 1)")
print("="*50)
print(f"   Charge: {charge}")
print(f"   Multiplicity: {mult_singlet}")
print(f"   Unpaired electrons: 0")
print(f"   Electronic state: Excited")
print(f"   Magnetic: NO (diamagnetic)")

# Triplet configuration
mult_triplet = 3
print("\n3️⃣  TRIPLET O₂ (Multiplicity = 3)")
print("="*50)
print(f"   Charge: {charge}")
print(f"   Multiplicity: {mult_triplet}")
print(f"   Unpaired electrons: 2")
print(f"   Electronic state: Ground")
print(f"   Magnetic: YES (paramagnetic)")

print("\n💡 Prediction: Triplet should be lower in energy!")
```

### ✅ Checkpoint 1

**Key concepts:**
- Multiplicity = 2S + 1
- Singlet (M=1): All paired
- Triplet (M=3): 2 unpaired
- Ground state O₂ is triplet (experimental fact)

**Question:**
- Why can't RHF (Restricted HF) handle triplet O₂?
- Answer: RHF forces pairing - need UHF (Unrestricted HF)!

## Step 5: Method Selection - RHF vs. UHF

### RHF (Restricted Hartree-Fock)
- Forces all electrons to pair
- Good for closed-shell molecules (H₂O, CH₄)
- **Cannot** describe unpaired electrons correctly

### UHF (Unrestricted Hartree-Fock)
- Allows different spatial orbitals for α and β spins
- Can handle open-shell systems
- **Required** for triplet O₂

### Rule of Thumb
- **Multiplicity = 1** → Use RHF
- **Multiplicity > 1** → Use UHF

### Run the cell below ⬇️

```python
# Method selection based on multiplicity
method_singlet = "RHF"  # Closed-shell → RHF
method_triplet = "UHF"  # Open-shell → UHF

basis = "6-31G"  # Use same basis for fair comparison

print("⚙️  METHOD SELECTION")
print("="*60)
print(f"Singlet O₂: {method_singlet}/{basis}")
print(f"   • Multiplicity = 1 → closed shell → RHF")
print(f"   • All electrons forced to pair")

print(f"\nTriplet O₂: {method_triplet}/{basis}")
print(f"   • Multiplicity = 3 → open shell → UHF")
print(f"   • Allows 2 unpaired electrons")

print("\n💡 Using different methods for fair comparison")
print("="*60)
```

### ✅ Checkpoint 2

**Method selection rules:**
- Singlet (M=1) → RHF ✓
- Doublet (M=2) → UHF ✓
- Triplet (M=3) → UHF ✓
- General: M > 1 needs UHF (or ROHF)

## Step 9: Spin Contamination (Advanced)

For UHF calculations, check spin contamination.

### What is Spin Contamination?
- UHF allows α and β orbitals to differ
- Can mix in higher spin states
- Pure triplet: $\langle S^2 \rangle = S(S+1) = 2.0$
- Contaminated: $\langle S^2 \rangle > 2.0$

### Why It Matters
- Large contamination → unreliable results
- Indicator of multi-reference character
- May need better method (ROHF, CASSCF)

### General Rules

**Determining Multiplicity:**
1. Count total electrons
2. Fill lowest orbitals following Hund's rule
3. Count unpaired electrons ($n$)
4. $M = n + 1$

**Common Mistakes:**
- ❌ Using RHF for radicals
- ❌ Wrong multiplicity (singlet for O₂)
- ❌ Forgetting charge affects electrons

## 🚀 Practical Applications

### When Multiplicity Matters Most

**1. Transition metals:**
- Often have unpaired d-electrons
- High-spin vs. low-spin states
- Critical for catalysis, magnetism

**2. Radicals:**
- Reactive intermediates
- Atmospheric chemistry
- Combustion processes

**3. Excited states:**
- Singlet vs. triplet excited states
- Photochemistry
- Phosphorescence

**4. Carbenes:**
- CH₂ ground state is triplet
- Singlet is reactive form
- ~9 kcal/mol gap

### Examples to Try

1. **Hydroxyl radical (·OH):**
   - 9 electrons → doublet
   - Important in atmosphere

2. **Methylene (CH₂):**
   - Triplet ground state
   - Singlet excited state

3. **NO (nitric oxide):**
   - Odd-electron molecule
   - Doublet ground state

4. **O₃ (ozone):**
   - Singlet ground state
   - But different from O₂!
