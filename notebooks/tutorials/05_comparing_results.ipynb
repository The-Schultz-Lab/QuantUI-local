## Step 4: Create Comparison Tables

Organize results for easy comparison.

### Run the cell below ⬇️

```python
# Create comparison tables
print("📊 COMPARISON TABLES")
print("="*70)

if collected_data:
    # Table 1: Basis set comparison (if available)
    basis_comparison = df[df['job_name'].str.contains('basis', case=False)]
    if not basis_comparison.empty:
        print("\n1️⃣  BASIS SET COMPARISON (H₂O)")
        print("-"*70)
        basis_comp = basis_comparison[['basis', 'energy_ha', 'energy_eV']].sort_values('energy_ha')
        print(basis_comp.to_string(index=False))
        
        # Calculate energy spans
        if len(basis_comp) > 1:
            energy_range = basis_comp['energy_ha'].max() - basis_comp['energy_ha'].min()
            print(f"\n   Energy range: {energy_range:.6f} Ha = {energy_range*627.5:.2f} kcal/mol")
    
    # Table 2: Multiplicity comparison (if available)
    mult_comparison = df[df['job_name'].str.contains('o2', case=False)]
    if not mult_comparison.empty:
        print("\n2️⃣  MULTIPLICITY COMPARISON (O₂)")
        print("-"*70)
        mult_comp = mult_comparison[['molecule', 'method', 'energy_ha']].sort_values('energy_ha')
        print(mult_comp.to_string(index=False))
        
        # Calculate singlet-triplet gap
        if len(mult_comp) >= 2:
            energies = dict(zip(mult_comp['molecule'], mult_comp['energy_ha']))
            if 'singlet' in energies and 'triplet' in energies:
                gap = (energies['singlet'] - energies['triplet']) * 627.5
                print(f"\n   Singlet-Triplet gap: {gap:.2f} kcal/mol")
    
    # Table 3: Charged species comparison (if available)
    charge_comparison = df[df['job_name'].str.contains('charged', case=False)]
    if not charge_comparison.empty:
        print("\n3️⃣  CHARGED SPECIES COMPARISON (NH₃ series)")
        print("-"*70)
        charge_comp = charge_comparison[['molecule', 'energy_ha']].sort_values('energy_ha')
        print(charge_comp.to_string(index=False))
    
    # Table 4: Overall summary
    print("\n4️⃣  OVERALL SUMMARY")
    print("-"*70)
    summary = df[['job_name', 'method', 'basis', 'energy_ha']]
    print(summary.to_string(index=False))
    
else:
    print("⚠️  No data available for comparison tables")
    print("   Complete Tutorials 1-4 first")

print("\n" + "="*70)
```

## Step 5: Statistical Analysis

Calculate statistics across your results.

### Run the cell below ⬇️

```python
# Statistical analysis
print("📈 STATISTICAL ANALYSIS")
print("="*70)

if collected_data:
    # Group by basis set
    if 'basis' in df.columns and len(df['basis'].unique()) > 1:
        print("\n1️⃣  Energy Statistics by Basis Set")
        print("-"*70)
        basis_stats = df.groupby('basis')['energy_ha'].agg(['count', 'mean', 'std', 'min', 'max'])
        print(basis_stats)
    
    # Group by method
    if 'method' in df.columns and len(df['method'].unique()) > 1:
        print("\n2️⃣  Energy Statistics by Method")
        print("-"*70)
        method_stats = df.groupby('method')['energy_ha'].agg(['count', 'mean', 'std'])
        print(method_stats)
    
    # Energy distribution
    print("\n3️⃣  Overall Energy Distribution")
    print("-"*70)
    print(f"   Total calculations: {len(df)}")
    print(f"   Mean energy: {df['energy_ha'].mean():.6f} Ha")
    print(f"   Std deviation: {df['energy_ha'].std():.6f} Ha")
    print(f"   Min energy: {df['energy_ha'].min():.6f} Ha")
    print(f"   Max energy: {df['energy_ha'].max():.6f} Ha")
    print(f"   Energy range: {df['energy_ha'].max() - df['energy_ha'].min():.6f} Ha")
    
else:
    print("⚠️  No data for statistical analysis")

print("\n" + "="*70)
```

## Step 6: Visualization

Create plots to visualize trends.

### Run the cell below ⬇️

```python
# Create visualizations
try:
    import matplotlib.pyplot as plt
    import numpy as np
    
    if collected_data and len(collected_data) >= 3:
        fig, axes = plt.subplots(1, 2, figsize=(14, 5))
        
        # Plot 1: Energy by Basis Set (if data available)
        basis_data = df[df['job_name'].str.contains('basis', case=False)]
        if not basis_data.empty and len(basis_data) >= 2:
            ax1 = axes[0]
            basis_sorted = basis_data.sort_values('energy_ha')
            colors = plt.cm.viridis(np.linspace(0, 1, len(basis_sorted)))
            
            bars = ax1.bar(range(len(basis_sorted)), basis_sorted['energy_ha'], 
                          color=colors, edgecolor='black', linewidth=1.5)
            ax1.set_xticks(range(len(basis_sorted)))
            ax1.set_xticklabels(basis_sorted['basis'], rotation=45)
            ax1.set_ylabel('Energy (Hartrees)', fontsize=12)
            ax1.set_title('Basis Set Convergence', fontsize=14, fontweight='bold')
            ax1.grid(axis='y', alpha=0.3)
            
            # Add value labels
            for i, (bar, energy) in enumerate(zip(bars, basis_sorted['energy_ha'])):
                ax1.text(bar.get_x() + bar.get_width()/2., bar.get_height(),
                        f'{energy:.4f}', ha='center', va='bottom', fontsize=9)
        else:
            axes[0].text(0.5, 0.5, 'No basis set\ncomparison data', 
                        ha='center', va='center', fontsize=14)
            axes[0].set_xlim(0, 1)
            axes[0].set_ylim(0, 1)
            axes[0].axis('off')
        
        # Plot 2: Method/System Comparison
        ax2 = axes[1]
        if len(df) >= 3:
            # Group by molecule type
            df_plot = df.sort_values('energy_ha')
            colors = plt.cm.plasma(np.linspace(0, 1, len(df_plot)))
            
            bars = ax2.barh(range(len(df_plot)), df_plot['energy_ha'], 
                           color=colors, edgecolor='black', linewidth=1.5)
            ax2.set_yticks(range(len(df_plot)))
            ax2.set_yticklabels(df_plot['job_name'], fontsize=8)
            ax2.set_xlabel('Energy (Hartrees)', fontsize=12)
            ax2.set_title('All Calculations Comparison', fontsize=14, fontweight='bold')
            ax2.grid(axis='x', alpha=0.3)
        else:
            ax2.text(0.5, 0.5, 'Need more\ncalculations', 
                    ha='center', va='center', fontsize=14)
            ax2.set_xlim(0, 1)
            ax2.set_ylim(0, 1)
            ax2.axis('off')
        
        plt.tight_layout()
        plt.show()
        
        print("✓ Visualization complete!")
        
    else:
        print("ℹ️  Need at least 3 calculations for meaningful plots")
        
except ImportError:
    print("ℹ️  matplotlib not available - install for visualizations")
except Exception as e:
    print(f"⚠️  Visualization error: {e}")
```

## 🚀 Advanced Analysis Ideas

### Further Exploration

1. **Energy Decomposition:**
   - Separate nuclear-nuclear, nuclear-electron, electron-electron
   - Analyze contribution of each term

2. **Geometry Optimization Trajectories:**
   - Track energy vs. optimization step
   - Visualize convergence path

3. **Basis Set Extrapolation:**
   - Fit energy vs. basis size
   - Extrapolate to complete basis set limit

4. **Method Comparison:**
   - HF vs. DFT (B3LYP, PBE)
