## Step 2: Understanding Charge and Electrons

### Electron Count Formula
$$\text{Electrons} = \sum Z_i - Q$$

Where:
- $Z_i$ = atomic numbers
- $Q$ = molecular charge

### Examples
| Species | Atoms | Nuclear charge | Molecular charge | Electrons |
|---------|-------|----------|----------|-----------|
| NH₃ | N(7) + 3H(1) | 10 | 0 | 10 |
| NH₄⁺ | N(7) + 4H(1) | 11 | +1 | 10 |
| NH₂⁻ | N(7) + 2H(1) | 9 | -1 | 10 |

**All have 10 electrons but different structures!**

### Run the cell below ⬇️

```python
# Count electrons for each species
species_info = {
    'NH3': {'atoms': {'N': 1, 'H': 3}, 'charge': 0},
    'NH4+': {'atoms': {'N': 1, 'H': 4}, 'charge': +1},
    'NH2-': {'atoms': {'N': 1, 'H': 2}, 'charge': -1},
}

print("ELECTRON COUNTING")
print("="*70)

atomic_numbers = {'N': 7, 'H': 1}

for name, info in species_info.items():
    nuclear_charge = sum(atomic_numbers[atom] * count 
                        for atom, count in info['atoms'].items())
    electrons = nuclear_charge - info['charge']
    
    formula = ''.join(f"{atom}{count if count > 1 else ''}" 
                     for atom, count in info['atoms'].items())
    
    print(f"\n{name:6s} ({formula}):")
    print(f"   Nuclear charge: {nuclear_charge}")
    print(f"   Molecular charge: {info['charge']:+d}")
    print(f"   Electrons: {electrons}")

print("\n" + "="*70)
print("💡 Notice: All have 10 electrons (isoelectronic)")
```

### ✅ Checkpoint 1

**Key concepts:**
- Charge affects electron count
- Positive charge → fewer electrons
- Negative charge → more electrons
- Isoelectronic species have same electron count

## Step 3: Define Geometries

We need different geometries for each species.

### NH₃ (Pyramidal)
- Trigonal pyramidal
- Lone pair on N
- H-N-H angle ~107°

### NH₄⁺ (Tetrahedral)
- Perfect tetrahedral
- No lone pair
- H-N-H angle 109.5°

### NH₂⁻ (Bent)
- Bent geometry
- 2 lone pairs on N
- H-N-H angle ~104°

### Run the cell below ⬇️

```python
# Define geometries (approximate, non-optimized)

# NH3 - pyramidal
nh3_xyz = """N   0.000   0.000   0.116
H   0.000   0.937  -0.272
H   0.812  -0.469  -0.272
H  -0.812  -0.469  -0.272"""

# NH4+ - tetrahedral
nh4_xyz = """N   0.000   0.000   0.000
H   0.529   0.529   0.529
H   0.529  -0.529  -0.529
H  -0.529   0.529  -0.529
H  -0.529  -0.529   0.529"""

# NH2- - bent
nh2_xyz = """N   0.000   0.000   0.143
H   0.000   0.805  -0.334
H   0.000  -0.805  -0.334"""

print("✓ Geometries defined for all three species")
print("\nNH₃  (neutral): 4 atoms, pyramidal")
print("NH₄⁺ (cation):  5 atoms, tetrahedral")
print("NH₂⁻ (anion):   3 atoms, bent")
```

## Step 4: Determine Multiplicity for Each Species

### NH₃ (10 electrons, charge=0)
- All electrons paired
- Multiplicity = 1 (singlet)

### NH₄⁺ (10 electrons, charge=+1)
- All electrons paired
- Multiplicity = 1 (singlet)

### NH₂⁻ (10 electrons, charge=-1)
- All electrons paired
- Multiplicity = 1 (singlet)

**All are singlets!** (Even number of electrons, all paired)

### Run the cell below ⬇️

```python
# Create molecule objects with correct charge and multiplicity

# NH3
atoms_nh3, coords_nh3 = parse_xyz_input(nh3_xyz)
mol_nh3 = Molecule(atoms_nh3, coords_nh3, charge=0, multiplicity=1)

# NH4+
atoms_nh4, coords_nh4 = parse_xyz_input(nh4_xyz)
mol_nh4 = Molecule(atoms_nh4, coords_nh4, charge=+1, multiplicity=1)

# NH2-
atoms_nh2, coords_nh2 = parse_xyz_input(nh2_xyz)
mol_nh2 = Molecule(atoms_nh2, coords_nh2, charge=-1, multiplicity=1)

print("MOLECULE SUMMARY")
print("="*70)

for name, mol in [('NH3', mol_nh3), ('NH4+', mol_nh4), ('NH2-', mol_nh2)]:
    print(f"\n{name}:")
    print(f"   Formula: {mol.get_formula()}")
    print(f"   Electrons: {mol.get_num_electrons()}")
    print(f"   Charge: {mol.charge:+d}")
    print(f"   Multiplicity: {mol.multiplicity}")

print("\n" + "="*70)
print("✓ All molecules validated and ready")
```

### ✅ Checkpoint 2

**Verify:**
- All have 10 electrons
- All have multiplicity = 1
- Charges: 0, +1, -1
- Different number of atoms

## Step 8: Calculate Ionization and Protonation Energies

Compute meaningful energy differences.

### Definitions
- **Ionization Energy (IE):** Energy to remove electron
  $$IE = E(\text{cation}) - E(\text{neutral})$$

- **Electron Affinity (EA):** Energy gain from adding electron
  $$EA = E(\text{neutral}) - E(\text{anion})$$

- **Proton Affinity (PA):** Energy of protonation
  $$PA = E(\text{NH}_3) - E(\text{NH}_4^+)$$

### Run the cell below ⬇️

```python
# Calculate derived quantities
print("🔬 ENERGY ANALYSIS")
print("="*70)

if len(nh3_results) == 3:
    E_nh3 = nh3_results['NH3']
    E_nh4 = nh3_results['NH4+']
    E_nh2 = nh3_results['NH2-']
    
    # Protonation: NH3 + H+ → NH4+
    # (We're comparing NH3 vs NH4+ without the proton energy)
    print("\n1️⃣  Protonation Energy (NH₃ → NH₄⁺)")
    print("-"*70)
    proton_energy = E_nh4 - E_nh3
    proton_energy_kcal = proton_energy * 627.5
    print(f"   E(NH₄⁺) - E(NH₃) = {proton_energy:.6f} Ha")
    print(f"                    = {proton_energy_kcal:.1f} kcal/mol")
    print(f"   💡 Negative = protonation is favorable")
    
    # Deprotonation: NH3 → NH2- + H+
    print("\n2️⃣  Deprotonation Energy (NH₃ → NH₂⁻)")
    print("-"*70)
    deproton_energy = E_nh2 - E_nh3
    deproton_energy_kcal = deproton_energy * 627.5
    print(f"   E(NH₂⁻) - E(NH₃) = {deproton_energy:.6f} Ha")
    print(f"                    = {deproton_energy_kcal:.1f} kcal/mol")
    print(f"   💡 Positive = deprotonation is unfavorable")
    
    # Relative stability
    print("\n3️⃣  Relative Stability (10-electron series)")
    print("-"*70)
    sorted_species = sorted(nh3_results.items(), key=lambda x: x[1])
    for i, (name, energy) in enumerate(sorted_species, 1):
        print(f"   {i}. {name:6s}: {energy:.6f} Ha (most stable)" if i == 1 
              else f"   {i}. {name:6s}: {energy:.6f} Ha")
    
    print("\n💡 Most negative energy = most stable")
    print("="*70)
    
else:
    print("⏳ Waiting for all calculations...")
```

### ✅ Checkpoint 3

**Expected observations:**
- NH₄⁺ typically most stable (most negative energy)
- NH₃ middle
- NH₂⁻ least stable (anions higher energy without diffuse functions)

**Key insights:**
- Protonation usually favorable (negative ΔE)
- Deprotonation usually unfavorable (positive ΔE)
- Charge affects energy significantly

### Key Concepts

**Charge and Electrons:**
- Cation (+): Fewer electrons
- Anion (-): More electrons
- Formula: $e^- = \sum Z_i - Q$

**Multiplicity Rules:**
- Even electrons → usually singlet
- Odd electrons → doublet (or higher)
- Exceptions exist!

**Energy Comparisons:**
- Cannot compare different atom counts directly
- Use: ionization, protonation, electron affinity
- Include all species in reactions

**Basis Sets for Anions:**
- Need diffuse functions (+ or ++)
- 6-31+G, 6-311++G** recommended
- Without diffuse: anions poorly described

### Practical Tips

**Common Charged Species:**
- **Acids:** Compare HA vs A⁻
- **Bases:** Compare B vs BH⁺
- **Radicals:** Often charged intermediates
- **Metal ions:** Many oxidation states
